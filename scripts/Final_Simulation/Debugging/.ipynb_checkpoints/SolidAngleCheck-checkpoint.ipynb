Just checking solid angle function (mathematically checking and plotting)

```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
from RayTraceFunctions import * 
from random import uniform
import random
import pickle
import math
import time
```

```python
thetG = [0,0,0]
spo = [-160.375,-113.,-80.0] #global CENTERED
n = 100
r = 0

#lim = np.pi/6 #30 degrees
#lim= np.pi/9 #20 degrees
#lim = np.pi/18 #10 degrees

Rays_NoLim = makerays_Zero(spo,thetG,r,n)

Rays_10 = makerays_Zero_narrow(spo,thetG,r,n,np.pi/18)
Rays_20 = makerays_Zero_narrow(spo,thetG,r,n,np.pi/9)
Rays_30 = makerays_Zero_narrow(spo,thetG,r,n,np.pi/6)

pl3dsurf = pl.figure().gca(projection = '3d')

Rays_NoLim_Theta = []
Rays_10_Theta = []
Rays_20_Theta = []
Rays_30_Theta = []
for i in range (0,int(len(Rays_NoLim))):
    x,y,z = ML(Rays_NoLim[i][2], Rays_NoLim[i][3], 263/2)
    pl3dsurf.plot(x,y,z, color = 'black',linewidth = .5)
    v = Rays_NoLim[i][3]
    x1,y1,z1 = v[0],v[1],v[2]
    theta = np.pi/2 -np.arctan(z1/np.sqrt((x1**2)+(y1**2)))
    Rays_NoLim_Theta.append(theta*180/np.pi)
    
for i in range (0,int(len(Rays_10))):
    x,y,z = ML(Rays_10[i][2], Rays_10[i][3], 263/2)
    pl3dsurf.plot(x,y,z, color = 'red',linewidth = .5)
    v = Rays_10[i][3]
    x1,y1,z1 = v[0],v[1],v[2]
    theta = np.pi/2 -np.arctan(z1/np.sqrt((x1**2)+(y1**2)))
    Rays_10_Theta.append(theta*180/np.pi)
    
for i in range (0,int(len(Rays_20))):
    x,y,z = ML(Rays_20[i][2], Rays_20[i][3], 263/2)
    pl3dsurf.plot(x,y,z, color = 'blue',linewidth = .5)
    v = Rays_20[i][3]
    x1,y1,z1 = v[0],v[1],v[2]
    theta = np.pi/2 -np.arctan(z1/np.sqrt((x1**2)+(y1**2)))
    Rays_20_Theta.append(theta*180/np.pi)
    
for i in range (0,int(len(Rays_30))):
    x,y,z = ML(Rays_30[i][2], Rays_30[i][3], 263/2)
    pl3dsurf.plot(x,y,z, color = 'green',linewidth = .5)
    v = Rays_30[i][3]
    x1,y1,z1 = v[0],v[1],v[2]
    theta = np.pi/2 -np.arctan(z1/np.sqrt((x1**2)+(y1**2)))
    Rays_30_Theta.append(theta*180/np.pi)
    
pl.show()

print(np.mean(Rays_NoLim_Theta))
print(np.mean(Rays_30_Theta))
print(np.mean(Rays_20_Theta))
print(np.mean(Rays_10_Theta))
```
