```python
def pos(A):
    B = []
    for i in range(len(A)):
        if A[i]>=0:
            B.append(A[i])
        else:
            B.append(-A[i])
    return B
            
```

```python
with open( 'Plots/new_polarizer_34.pkl' , 'rb') as file1:
    d=pickle.load(file1, encoding = 'latin1')
Power1 = (d['sig0F'])
Power = [x/Power1.max() for x in Power1]
Delay = (d['delay0F'])
pl.plot(Delay+4,Power,linewidth = 1)
pl.title('Physical FTS')
pl.xlabel('Optical Delay (mm)')
pl.xlim(-67,60)
pl.savefig('Plots/ACTUAL'+'90.3Ghz'+time.strftime("_%Y_%m_%d_%H_%M_%S")+'.png')
pl.show()
```
