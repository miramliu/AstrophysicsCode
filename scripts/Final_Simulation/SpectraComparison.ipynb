Comparing Physical FTS and Simulation Spectra for 90Ghz.

Approximation. simulation needed because can't actually do path integrals etc. 
how to learn about the properties of the instrument for what we need it for. 

resolution is two bins. which is what is expected with a hanning function. 
important to note why this is happening. 
the width of the spectrum is due to the hanning function which makes it a triangle. 
THIS is due to the fact that the path of the mirror is NOT PARALLEL to the light path. 
so when you image back it's kinda weird. 
- this effect is added FOR the compactness. 

weird lump thing is due to the LACK of apodization
- second harmonic?
- weird resolution? why the distance between the points is so weird.
