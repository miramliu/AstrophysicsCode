This is a notebook going over the basic deconvolution of an interferogram. What is important to note is that this is only relevant at high carrier speeds. This is also based on another code written as a more complex Tkinter file.

The convolution of the interferogram is evident from the asymmetry of the two peaks, and how the white light fringe is off center.

It also should be noted that this analysis is used assuming parameters are known from a test run. The analysis finding these parameters is not included in this code, though these estimations are used.

A brief summary of the deconvolution is shown below.

### Line by line break down of analysis in FTI.py
~~~~
Nsize = 2**i
~~~~

Sets the number of samples used to a factor of two for the best Fourier Transform results. 

~~~~
dt = 1/frequency 
~~~~

By definition the smallest segment of time is the inverse of the frequency of sampling.

~~~~
t = dt(Nsize) 
~~~~

This is the total time of the data set.

~~~~
x = v*t 
~~~~

this is the total distance travelled in the data set

~~~~
total_s = (d['samples requested'])
startpt = ((total_s - Nsize)/2) 
endpt = startpt + Nsize
~~~~

This selects white data bins to be selected such that the number of samples is Nsize, centered at the white light fringe.

~~~~
a = d['delay' + self.t]
b = a[startpt:endpt]
~~~~

This selects the approriate corresponding delay 

~~~~
F = (d['sig'+self.t])
A = np.hanning(Nsize)*F[startpt:endpt]
S = np.fft.rfft(A)
~~~~

This selects the data set requested and multiplies it by a hanning function. After the hanning function is applied a Fourier Transform is applied. 

~~~~
s = S[:-1]
u = np.abs(s)
dNu = 1/(Nsize*dx)
Nu = dNu*np.arange(Nsize/2)
NuFull = dNu*np.arange((Nsize/2) + 1)
~~~~

This takes the appropriate samples from the Fourier Transform, and the absolute value to get only the real aspect. It also switches to k-space coordinates

~~~~
center = (300*(Nu[0] + Nu[-1]))/2
top = max(np.abs(s))
~~~~

This gets the center and peak in k-space

~~~~
tau = .04 #ms
k_v = 1/((1/tau)+1j*2*np.pi*300*NuFull)
~~~~

$$k_v = \frac{1}{\frac{1}{\tau} + i(2\pi f)}$$
This is the introduction of the time constant, tau. The equation is the convolution kernel taking into account the delay in the circuit. It is what convolves with the input to give the pure output with no signal processing. 


~~~~
n = 3
sig = n/(tau*(2*np.pi))
~~~~

$$\sigma = \frac{n}{2\pi \tau}$$
This is a constant in k-space to maintain propper units, and is used to filter out high frequencies


~~~~
f = (1/(sig*np.sqrt(np.pi)))*np.exp(-(NuFull**2)/(sig**2))
~~~~

$$f = \frac{1}{\sigma\sqrt{\pi}}e^\frac{-f^2}{\sigma^2}$$
This is the appropriate gaussian that the signal is modulated by.


~~~~
B = f*np.fft.rfft(A)/k_v*tau
~~~~

This is the fourier transform of the interferogram multiplied by the gaussian and divided by the convolution kernel times the time constant. This is the deconvolution. It is adjusting the signal in k-space to yield a more accurate final output. 

~~~~
c = np.fft.irfft(B)
~~~~

This is the inverse fourier transoform of the deconvolved spectrum, yielding the deconvolved interferogram.

~~~~
g = (np.abs(np.fft.rfft(c))/k_v*tau)
~~~~

This is the real aspect of the normalized deconvolved spectrum. 

~~~~
L=np.sum(g)
~~~~

This is the integral of the deconvolved spectrum.

```python
import matplotlib.pyplot as pl
import pickle
import numpy as np
import sys
%matplotlib inline
```

```python
with open('../../data/raw_data/20170615_1730_20mms_reference.pkl', 'rb') as f:
    d = pickle.load(f, encoding='latin1') 
```

```python
i = 9
Nsize = 2**i
dt=(1/(d['sample freq'])) #period
T1=dt*(Nsize) #full period
v=(d['speed'])
X = v*T1 #full distance
dx = dt*v #smallest amount of distance travelled
total_t = (d['scan time']) #how long it ran
    
total_s = (d['samples requested']) #number of samples 
startpt = ((total_s - Nsize)/2) #starting point


endpt = startpt + Nsize #ending point


df = 1/T1
f = df*np.arange(Nsize/2)+df/2.0
fFull = df*np.arange((Nsize/2) + 1)+df/2.0

startpt = int(startpt)
endpt = int(endpt)
F = (d['sig0F']) 
D = F[startpt:endpt]
#D = np.flipud(D)

a = d['delay0F']/v
t = a[startpt:endpt]

pl.plot(t,D)
pl.title('Original Interferogram')
pl.xlabel('Time(s)')
pl.show()
```

```python
A = np.hanning(Nsize)*F[startpt:endpt]
S = np.fft.rfft(A)
s = S[:-1]
u = np.abs(s)
dNu = 1/(Nsize*dx)
Nu = dNu*np.arange(Nsize/2)
NuFull = dNu*np.arange((Nsize/2) + 1)
center = (300*(Nu[0] + Nu[-1]))/2
top = max(np.abs(s))
tau = 0.0009 #ms
k_v = 1/((1/tau)+1j*2*np.pi*300*NuFull)
            
            
n = 3 #greater number filters out more high frequencies
sig = n/(tau*(2*np.pi)) 
f = (1/(sig*np.sqrt(np.pi)))*np.exp(-(NuFull**2)/(sig**2)) #gaussian
B = f*np.fft.rfft(A)/k_v*tau  #fourier transform of interferogram/kernel
c = np.fft.irfft(B) #inverse fourier transform of deconvolved spectrum (deconvolved interferogram)
            
# normalize gaussian to get rid of noise (divide by integral of itself) (in B)
            
g = (np.abs(np.fft.rfft(c))/k_v*tau)[:2**7] #CHANGE depending on number of data points
L=np.sum(g) #integral of spectrum
            
xint, yint = t, (1/L)*c
xspec, yspec = 300*NuFull, np.abs(np.fft.rfft(c))/k_v*tau
```

```python
pl.plot(xspec,yspec, label = 'Spectra')
pl.plot(300*NuFull,np.abs(S)*.002, label = 'Original spectrum', color = 'green')
pl.title('Deconvolved spectrum')
pl.xlabel('GHz')
pl.legend()
pl.show()
```

```python
pl.plot(xint,yint, label = 'Deconvolved interferogram')
pl.plot(t,D*.0079, label = 'Original interferogram', color = 'green')
pl.title('Deconvolved interferogram')
pl.legend(fontsize = 'x-small')
pl.show()
```

Clearly this deconvolution is not perfect (it is not symmetric or centered perfectly about 0). This is because the parameters or the model used to find the parameters are not exact. However, this is a good general explanation of what deconvolution should be.
