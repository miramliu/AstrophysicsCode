This is a simple exploration of how to rotate vectors from one coordinate system to another. It will essentially create two 'origins'defined by two planes. It then translates one vector coming from one origin into an identical vector with reference to the second origin. Mira

```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
import math
from RayTraceFun import *
```

This is a basic explanation of the change of a vector from XYZ to UVW

$$e_x = [1 0 0]$$
$$e_y = [0 1 0]$$
$$e_z = [0 0 1]$$

So in the first coordinate system: 

$$\vec{v}= \begin{bmatrix}v_x&v_y&v_z\end{bmatrix}\cdot\begin{bmatrix} e_x\\e_y\\e_z\end{bmatrix}$$

$$\vec{v} = v_x\hat{e_x} + v_y\hat{e_y} + v_z\hat{e_z}$$


Using the orthonormal basis of the other coordinate system, and a vector $\vec{O}$ connecting the initial and new origins you can get the same vector in another coordinate system using a Rotation matrix R. 

$$basis: \begin{bmatrix}e_u\\e_v\\e_w\end{bmatrix}$$

$$\vec{O} = \begin{bmatrix}o_u\\o_v\\o_w\end{bmatrix}$$

$$R = \hat{e_u}\hat{e_x} + \hat{e_v}\hat{e_y} + \hat{e_w}\hat{e_z}$$

So, using this the vector in the alternate coordinate system is given by the equation below.

$$\vec{v_f} = R\vec{v_i} + \vec{O}$$

$$\vec{v} = \begin{bmatrix}v_u\\v_v\\v_w\end{bmatrix} = R\begin{bmatrix}v_x\\v_y\\v_z\end{bmatrix} + \begin{bmatrix}o_u\\o_v\\o_w\end{bmatrix}$$

```python
#Rotation matrices around a given axis (in this case x,y,z)
#variable given is the angle of rotation wanted 
def Rx(x):
    Rx = np.matrix([[1,0,0],[0, np.cos(x), -np.sin(x)], [0, np.sin(x), np.cos(x)]])
    return Rx

def Ry(y):
    Ry = np.matrix([[np.cos(y),0,np.sin(y)],[0, 1, 0], [-np.sin(y), 0, np.cos(y)]])
    return Ry

def Rz(z):
    Rz = np.matrix([[np.cos(z), - np.sin(z), 0],[np.sin(z), np.cos(z), 0], [0, 0, 1]])
    return Rz
```

```python
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
v = [1,1,1]
for i in range (1,50):
    v2 = np.array(np.dot(v,Rx(0 + i)))
    x2,y2,z2 = make_line(origin,v2[0],10)
    pl.plot(x2,y2,z2) #rotated around the x-axis
pl.title('x rotation')
pl.show()
```

```python
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
v = [1,1,1]
for i in range (1,50):
    v2 = np.array(np.dot(v,Ry(0 + i)))
    x2,y2,z2 = make_line(origin,v2[0],10)
    pl.plot(x2,y2,z2) #rotated around the x-axis
pl.title('y rotation')
pl.show()
```

```python
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
v = [1,1,1]
for i in range (1,50):
    v2 = np.array(np.dot(v,Rz(0 + i)))
    x2,y2,z2 = make_line(origin,v2[0],10)
    pl.plot(x2,y2,z2) #rotated around the x-axis
pl.title('z rotation')
pl.show()
```

```python
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
origin2 = [2,1,1]
v = [1,1,1]
x,y,z = make_line(origin, v, 10)
v2 = np.array(np.dot(v,Rz(np.pi/2)))
#print(v2)
x2,y2,z2 = make_line(origin,v2[0], 10) #rotated pi/2 around z axis
x3,y3,z3 = make_line(origin2, v2[0], 10) #shifted origin and rotated pi/2
pl.plot(x,y,z)
pl.plot(x2,y2,z2)
pl.plot(x3,y3,z3)
pl.show()
```

```python
#translate and reverse translate and make sure to get the same line!
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
origin2 = [2,1,1]
v = [1,1,1]
v2 = np.array(np.dot(v,Rz(np.pi/2))) 
x,y,z = make_line(origin, v, 10) #original
x3,y3,z3 = make_line(origin2, v2[0], 10) #shifted origin and rotated pi/2
v3 = np.array(np.dot(v2,Rz(-np.pi/2))) 
x4,y4,z4 = make_line(origin,v3[0], 10 )
pl.plot(x,y,z, 'red') #original
pl.plot(x3,y3,z3, 'green') #shifted and rotated
pl.plot(x4,y4,z4, 'blue') #shifted and rotated back!
pl.show()
```
