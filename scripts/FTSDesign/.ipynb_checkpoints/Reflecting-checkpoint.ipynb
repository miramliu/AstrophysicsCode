```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
```

```python

#plot a line in 3d space? (incident ray)
#vector direction
d1 = np.array([.5,1,1])
#point it passes through
point2 = np.array([1,1,1])

xLi = range(10)
yLi = ((xL - point2[0])/(d1[0]))*d1[1] + point2[1]
zLi = ((xL - point2[0])/(d1[0]))*d1[2] + point2[2]
```

```python
coeff1 = np.array([1,1,1,1]) #a,b,c,d for plane
#gradient of plane
x,y,z = np.meshgrid(range(10), range(10), range(10))
U = coeff1[0]*x + coeff1[1]*y + coeff1[2]*z
dU = coeff1[:-1] #gradient (for plane)

#print (dU)
#plane
xp, yp = np.meshgrid(range(10), range(10))
zp = (-coeff1[0]*xp - coeff1[1]*yp + coeff1[3])/coeff1[2]

#normalize normal line
UNorm = np.sqrt((coeff1[0])**2 + (coeff1[1])**2 + (coeff1[2])**2)
N = ([u/UNorm for u in dU])


#plot of normal line?
point2 = np.array([0,0,0]) #x_0,y_0,z_0 (for where to intersect)
t = (range(10))
xL = point2[0] + t*coeff1[0]
yL = point2[1] + t*coeff1[1]
zL = point2[2] + t*coeff1[2]


#angle of intersection?
A1 = np.abs(coeff1[0]*dU[0] + coeff1[1]*dU[1] + coeff1[2]*dU[2])
A2 = np.sqrt(coeff1[0]**2 + coeff1[1]**2 + coeff1[2]**2)
A3 = np.sqrt(dU[0]**2 + dU[1]**2 + dU[2]**2)
A4 = A1/(A2*A3)
#print (A4) #so it is the normal line. 
#alpha = np.arcsin(A4)
#print ('incident angle in radians:', alpha)
#print ('incident angle in degrees:', alpha*57.3)


print ('normalized: ', N)
```

```python
#plot a line in 3d space (reflected ray)
#normalize d1
#d1Norm = np.sqrt((d1[0])**2 + (d1[1])**2 + (d1[2])**2)
#DN = np.dot(d1,N) #dot product of d and n
N = np.array(N)
d2 = d1 - 2*N

xLr = range(10)
yLr = ((xL - point2[0])/(d2[0]))*d2[1] + point2[1]
zLr = ((xL - point2[0])/(d2[0]))*d2[2] + point2[2]
```

```python
#plot a 3d surface with normal, incident ray, and reflected ray

pl3d = pl.figure().gca(projection = '3d')
#pl3d.set_xlim(left = 10, right = 0)
#pl3d.set_ylim(bottom = 0, top = 10)
#pl3d.set_zlim(bottom = 0, top = 2)
pl3d.plot_surface(xp,yp,zp, alpha = .2)

pl3d.plot(xL,yL,zL, 'black')
pl3d.plot(xLi,yLi,zLi, 'red')
pl3d.plot(xLr,yLr,zLr, 'blue')



ax=pl.gca()
ax.hold(True)
pl.show()
```
