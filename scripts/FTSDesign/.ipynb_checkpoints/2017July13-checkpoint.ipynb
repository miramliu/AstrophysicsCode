This is a plot of the ellipse (mirror in the FTS) that will be used in the ray tracer. The information regarding the legnths of the sides is gotten from the diagram in page 78 of the lab notebook.

It is an ellipse with an origin located between the two foci. The line that the foci lie on is the axis of rotation that the ellipse is then rotated around. 

The units here are assumed to be centimeters.

Mira

```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
import math
from mpl_toolkits.mplot3d import Axes3D
```

The equation of an ellipse is given by $\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$

In this ellipse (pre-rotation) b = 256, a = 263.9, c = 64.15

A graph of this ellipse is shown below


```python
pl3d = pl.figure().gca(projection = '3d')
origin = [0,0,0]
coeffellipse = [263.9,256,64.15] #a,b,c
xc=np.linspace(-coeffellipse[0],coeffellipse[0],100)
yc1 = np.sqrt((1-(xc**2)/(coeffellipse[0]**2))*coeffellipse[1])
yc2 = -np.sqrt((1-(xc**2)/(coeffellipse[0]**2))*coeffellipse[1])
zc = np.linspace(0,0,100)
pl.plot(xc,yc1,zc, 'blue')
#pl.plot(xc,yc2,zc, 'blue')
pl.show()
```

This ellipse is then rotated around the x-axis.

```python
def Rx(x):
    Rx = np.matrix([[1,0,0],[0, np.cos(x), -np.sin(x)], [0, np.sin(x), np.cos(x)]])
    return Rx
```

```python
Rotated = []
for i in range (0,100):
    v = [xc[i], yc1[i],zc[i]]
    v2 = np.array(np.dot(v,Rx(0 + i)))
    Rotated.append(v2[0])
xcR = []
ycR = []
zcR = []
for j in range (0,100):
    xcR.append(Rotated[j][0])
    ycR.append(Rotated[j][1])
    zcR.append(Rotated[j][2])
```

```python
#how this is rotated around the x axis
pl3d = pl.figure().gca(projection = '3d')
pl.plot(xcR,ycR,zcR, 'blue')
pl.show()
```

```python
Rotated = []
for i in range (0,100): 
    v = [xc[i], yc1[i],zc[i]] # number of original points
    v2 = np.array(np.dot(v,Rx(np.pi/2))) #multiplied by rotation vector
    Rotated.append(v2[0]) #rotated vectors
xcR1 = []
ycR1 = []
zcR1 = []
for j in range (0,100):
    xcR1.append(Rotated[j][0])
    ycR1.append(Rotated[j][1])
    zcR1.append(Rotated[j][2])
    
```

```python
def createstrand(theta):
    Rotated = []
    for i in range (0,100): 
        v = [xc[i], yc1[i],zc[i]] # number of original points
        v2 = np.array(np.dot(v,Rx(theta))) #multiplied by rotation vector
        Rotated.append(v2[0]) #rotated vectors
    xcR1 = []
    ycR1 = []
    zcR1 = []
    for j in range (0,100):
        xcR1.append(Rotated[j][0])
        ycR1.append(Rotated[j][1])
        zcR1.append(Rotated[j][2])
    return xcR1,ycR1,zcR1
```

```python
#how this is rotated around the x axis by pi/2
pl3d = pl.figure().gca(projection = '3d')
pl.plot(xc,yc1,zc, 'blue')
x,y,z = createstrand(np.pi/2)
pl.plot(x,y,z ,'green')
pl.show()
```

```python
x,y,z = createstrand(np.pi/4)
```

```python
def createz(a):
    x1 = []
    y1 = []
    z1 = []
    for i in range (0,int(a)):
        theta = i*np.pi/20
        x,y,z = createstrand(theta)
        x1.extend(x)
        y1.extend(y)
        z1.extend(z)
    return x1,y1,z1
```

```python
pl3dsurf = pl.figure().gca(projection = '3d')
X,Y,Z = createz(10)
X.pop(0)
Y.pop(0)
Z.pop(0)
pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2)
#pl3dsurf.set_zlim(-4,0)
pl.show()
```
