```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
import math
```

```python
coeff3 = np.array([0, 0, 3])
xpoint = []
ypoint = []
for x in range(coeff3[0]-coeff3[2],coeff3[0]+coeff3[2]):
    for y in range(coeff3[1]-coeff3[2],coeff3[1]+coeff3[2]):
        d = np.sqrt(((x-coeff3[0])**2) + ((y - coeff3[1])**2))
        if d <= coeff3[2]: 
            xpoint.append(x)
            ypoint.append(y)
```

```python
#these are the points the rays must go through all on xy plane
points = []
for i in range(0,len(xpoint)-1):
    points.append([xpoint[i],ypoint[i], 0])
#print (points)
points3 = np.array(points[1])
print ('These are the coordinates: ', points3)
```

```python
#making the rays from each point and finite length
pl3d = pl.figure().gca(projection = '3d')
#pl3d.set_xlim(left = -10, right = 10)
#pl3d.set_ylim(bottom = -10, top = 10)
pl3d.set_zlim(bottom=0, top = 10)

#incident ray
VectL = np.array([.01,1,1])
VNorm = np.sqrt(VectL[0]**2 + VectL[1]**2 + VectL[2]**2)
VectLNorm = ([u/VNorm for u in VectL])
VectLNorm = np.array(VectLNorm)#a2,b2,c2 for line

#making the plane
coeff1 = np.array([.01,.01,1,1]) #a1,b1,c1,d for plane
#gradient of plane
xpl,ypl,zpl = np.meshgrid(range(10), range(10), range(10))
U = coeff1[0]*xpl + coeff1[1]*ypl + coeff1[2]*zpl
dU = coeff1[:-1] #gradient (for plane)

#normalized normal line
UNorm = np.sqrt((coeff1[0])**2 + (coeff1[1])**2 + (coeff1[2])**2)
N = ([u/UNorm for u in dU])
N = np.array(N)

#reflected ray
VectL2 = VectL + 2*N
VNorm2 = np.sqrt(VectL2[0]**2 + VectL2[1]**2 + VectL2[2]**2)
VectLNorm2 = ([u/VNorm2 for u in VectL2])
VectLNorm2 = np.array(VectLNorm2)#a2,b2,c2 for reflected ray

#print (dU)
#plane
xp, yp = np.meshgrid(range(-10,10), range(-10,10))
zp = (-coeff1[0]*xp - coeff1[1]*yp + coeff1[3])/coeff1[2]

for i in range (0,len(points)):
    #point2 = points[i] #x_0,y_0,z_0 ORIGINAL POINT FOR INCIDENT LINES
    point2 = ([0,0,0]) 
    #incident
    x1 = []
    y1 = []
    z1 = []
    #reflected
    x2 = []
    y2 = []
    z2 = []
    #reflected ray
    VectL2 = VectL + 2*N
    VNorm2 = np.sqrt(VectL2[0]**2 + VectL2[1]**2 + VectL2[2]**2)
    VectLNorm2 = ([u/VNorm2 for u in VectL2])
    VectLNorm2 = np.array(VectLNorm2)#a2,b2,c2 for reflected ray
    #intersection for plot
    Cb = (coeff1[1]*VectLNorm[1])/VectLNorm[0]
    Cc = (coeff1[2]*VectLNorm[2])/VectLNorm[0]
    denom = coeff1[0] + Cb + Cc
    xint = (coeff1[3] - point2[2] - point2[1] + Cb*point2[0] + Cc*point2[0])/denom
    yint = ((xint - point2[0])/(VectLNorm[0]))*VectLNorm[1] + point2[1]
    zint = ((xint - point2[0])/(VectLNorm[0]))*VectLNorm[2] + point2[2]
    pointint = []
    pointint.append(xint) #where line intersects
    pointint.append(yint)
    pointint.append(zint)
    #incident
    for t in range (0,20):
        L = np.sqrt(((VectL[0]*t)**2 + (VectL[2]*t)**2 + (VectL[2]*t)**2))
        xL = pointint[0] + t*VectLNorm[0]
        yL = pointint[1] + t*VectLNorm[1]
        zL = pointint[2] + t*VectLNorm[2]
        if L <= 20:
            x1.append(xL)
            y1.append(yL)
            z1.append(zL)
    #reflected
    for t in range (0,20):
        L2 = np.sqrt(((VectL2[0]*t)**2 + (VectL2[2]*t)**2 + (VectL2[2]*t)**2))
        xL2 = pointint[0] + t*VectLNorm2[0]
        yL2 = pointint[1] + t*VectLNorm2[1]
        zL2 = pointint[2] + t*VectLNorm2[2]
        if L2 <= 20:
            x2.append(xL2)
            y2.append(yL2)
            z2.append(zL2)
pl3d.plot(x1,y1,z1, 'blue') 
pl3d.plot(x2,y2,z2, 'red')




pl3d.plot_surface(xp,yp,zp, alpha = .2)

   
ax=pl.gca()
ax.hold(True)
pl.show()
```

```python
def R_theta(axis,theta):
    axis = np.asarray(axis)
    axis = axis/math.sqrt(np.dot(axis,axis))
    a = math.cos(theta/2.0)
    b,c,d = -axis*math.sin(theta/2.0)
    aa,bb,cc,dd = a*a, b*b, c*c,d*d
    bc,ad,ac,ab,bd,cd = b*c,a*d,a*c,a*b,b*d,c*d
    return np.array([[aa+bb-cc-dd, 2*(bc+ad), 2*(bd-ac)],
                     [2*(bc-ad), aa+cc-bb-dd, 2*(cd+ab)],
                     [2*(bd+ac), 2*(cd-ab), aa+dd-bb-cc]])
v = [3, 5, 0]
axis = [4, 4, 1]
theta = 1.2 
print(np.dot(R_theta(axis,theta), v)) 
```

```python
#plotting rays
pl3d = pl.figure().gca(projection = '3d')
#ray 1 
V = [1,1,1]
axis1 = [0,0,6]
theta1 = np.pi/2.0
x1 = []
y1 = []
z1 = []
pointint = [0,0,0]
#original
for t in range (0,20):
    L = np.sqrt(((V[0]*t)**2 + (V[2]*t)**2 + (V[2]*t)**2))
    xL = pointint[0] + t*V[0]
    yL = pointint[1] + t*V[1]
    zL = pointint[2] + t*V[2]
    if L <= 20:
            x1.append(xL)
            y1.append(yL)
            z1.append(zL)
pl3d.plot(x1,y1,z1, 'blue')

V2 = np.array((np.dot(R_theta(axis1,theta1), V)) )

#rotated
x2 = []
y2 = []
z2 = []
for t in range (0,20):
    L = np.sqrt(((V2[0]*t)**2 + (V2[2]*t)**2 + (V2[2]*t)**2))
    xL = pointint[0] + t*V2[0]
    yL = pointint[1] + t*V2[1]
    zL = pointint[2] + t*V2[2]
    if L <= 20:
            x2.append(xL)
            y2.append(yL)
            z2.append(zL)
pl3d.plot(x2,y2,z2, 'red')

#plot axis of rotation
x3 = []
y3 = []
z3 = []
for t in range (0,20):
    L = np.sqrt(((axis1[0]*t)**2 + (axis1[2]*t)**2 + (axis1[2]*t)**2))
    xL = pointint[0] + t*axis1[0]
    yL = pointint[1] + t*axis1[1]
    zL = pointint[2] + t*axis1[2]
    if L <= 20:
            x3.append(xL)
            y3.append(yL)
            z3.append(zL)
pl3d.plot(x3,y3,z3, 'black')


ax=pl.gca()
ax.hold(True)
pl.show()
```
