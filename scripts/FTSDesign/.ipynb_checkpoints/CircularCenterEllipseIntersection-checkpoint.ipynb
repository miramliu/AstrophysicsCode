```python
def setrange2d(xrange,X,Y,Z, origin): #given circle and intersection points, only keep points within CIRCLE
    xintG = []
    yintG = []
    zintG = []
    for i in range (0,len(X)):
        xinti = X[i]
        yinti = Y[i]
        zinti = Z[i]
        if (xinti-origin[0])**2 + (zinti)**2 <= xrange**2:
            xintG.append(xinti)
            yintG.append(yinti)
            zintG.append(zinti)
    return xintG,yintG,zintG
#now make a circle: (x-h)^2 + (y-k)^2 = r^2
#origin is 0,0
def setrange3d(xrange,X,Y,Z,origin):
    xint3d = []
    yint3d = []
    zint3d = []
    Goodi = []
    for i in range (0,len(X)):
        xinti = X[i]
        yinti = Y[i]
        zinti = Z[i]
        if (xinti-origin[0])**2 + (zinti)**2 <= xrange**2 and yinti >= 0:
            xint3d.append(xinti)
            yint3d.append(yinti)
            zint3d.append(zinti)
            Goodi.append(i)
    return xint3d,yint3d,zint3d, Goodi
```

```python
#EDITTED SPECIFICALLY TO MAKE INTERSECTION CLEAR
def reflectellipsePOINT(coeffellipse,r,v,p,L):
        c = tangcoef(p,v,coeffellipse) #plane coefficients
        VectL = v #incident vector#defining points of incident vector
        VectLinit = [-a for a in v]
        VectLNorm = norm(v) #incident unit vector
        dU, N = plane_info(c,r) #gradient and normal of plane
        #reflected ray
        #VectL2 = VectLNorm + 2*N #reflected vector
        VectL2 = VectLNorm - 2*N #IS IT PLUS OR MINUS
        VectLNorm2 = norm(VectL2) #reflected unit vector
        xp,yp,zp = make_plane(c,r) #plane
        xint,yint,zint = ellipselineint(p,v,coeffellipse)
        #pointintRETURN = [xint,yint,zint]
        pointint = [float(xint[0]),float(yint[0]),float(zint[0])] #array and points of intersection
        pointintneg = [float(xint[1]),float(yint[1]),float(zint[1])] #array and points of intersection
        xi,yi,zi = make_line(pointint,VectLinit,L) #incident line from intersection point
        xr,yr,zr = make_line(pointint,VectL2,L)
        xin,yin,zin = make_line(pointintneg,VectLinit,L) #incident line from intersection point
        xrn,yrn,zrn = make_line(pointintneg,VectL2,L)
        xi.extend(xin),yi.extend(yin), zi.extend(zin), xr.extend(xrn), yr.extend(yrn), zr.extend(zrn)
        return xi,yi,zi,xr,yr,zr,xp,yp,zp, pointint, pointintneg
#Reflection off of ellipse with a specular source FROM ONE POINT (in this case referred to as 'origin')
def reflect_specORIGINellipsePOINT(coeffellipse,r,origin,L, theta, specnum):
    xi = []
    yi = []
    zi = []
    xr = []
    yr = []
    zr = []
    xp = []
    yp = []
    zp = []
    pointints = []
    xspec,yspec,zspec = spec(specnum)
    for i in range (1, specnum-1):
        Vi = [xspec[i], yspec[i], zspec[i]]
        Vi2 = np.array(np.dot(Vi,Rx(theta)))
        xiL,yiL,ziL,xrL,yrL,zrL,xpL,ypL,zpL,pointint, pointintneg = reflectellipsePOINT(coeffellipse,r,Vi2[0],origin,L)
        xi.append(xiL)
        yi.append(yiL)
        zi.append(ziL)
        xr.append(xrL)
        yr.append(yrL)
        zr.append(zrL)
        xp.append(xpL)
        yp.append(ypL)
        zp.append(zpL)
        pointints.append(pointint)
        pointints.append(pointintneg)
    return xi,yi,zi,xr,yr,zr,xp,yp,zp, pointints
```

```python
def getGoodi(points, xrange):
    xintall1, yintall1, zintall1 = getpoints(points)
    xint1, yint1, zint1, Goodi1 = setrange3d(xrange,xintall1, yintall1, zintall1, origin)
    return Goodi1

def selectrays(xi,yi,zi,xr,yr,zr,xp,yp,zp, pointints,pointsintsneg, xrange, origin):
    xiG = []
    yiG = []
    ziG = []
    xrG = []
    yrG = []
    zrG = []
    xpG = []
    ypG = []
    zpG = []
    Goodis = []
    Goodi1 = getGoodi(pointints, xrange)
    Goodi2 = getGoodi(pointsintsneg, xrange)
    for i in Goodi1:
        xiG.append(xi[i])
        yiG.append(yi[i])
        ziG.append(zi[i])
        xrG.append(xr[i])
        yrG.append(yr[i])
        zrG.append(zr[i])
        xpG.append(xp[i])
        ypG.append(yp[i])
        zpG.append(xp[i])
    for f in Goodi2:
        xiG.append(xi[f])
        yiG.append(yi[f])
        ziG.append(zi[f])
        xrG.append(xr[f])
        yrG.append(yr[f])
        zrG.append(zr[f])
        xpG.append(xp[f])
        ypG.append(yp[f])
        zpG.append(xp[f])
    return xiG,yiG,ziG,xrG,yrG,zrG,xpG,ypG,zpG
```
