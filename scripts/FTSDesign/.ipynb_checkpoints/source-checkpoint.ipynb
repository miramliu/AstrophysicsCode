```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
```

```python
coeff3 = np.array([0, 0, 3])
xpoint = []
ypoint = []
for x in range(coeff3[0]-coeff3[2],coeff3[0]+coeff3[2]):
    for y in range(coeff3[1]-coeff3[2],coeff3[1]+coeff3[2]):
        d = np.sqrt(((x-coeff3[0])**2) + ((y - coeff3[1])**2))
        if d <= coeff3[2]: 
            xpoint.append(x)
            ypoint.append(y)
            #print (x,y)
#this is the points that are in the circle

#getting the points of intersection in a circle
#NOW SUPERIMPOSING CIRCLE AND THE POINTS IN THE CIRCLE? 
#making circular plane
coeff3 = np.array([0, 0, 3]) #h,k,r for circle
xc=np.linspace(coeff3[0]-coeff3[2],coeff3[0]+coeff3[2],100)
yc1 = np.sqrt((coeff3[2]**2)-((xc-coeff3[0])**2)) + coeff3[1]
yc2 = -np.sqrt((coeff3[2]**2)-((xc-coeff3[0])**2)) + coeff3[1]
zc = 0
pl3d = pl.figure().gca(projection = '3d')
#pl3d.plot_surface(xc,yc1,zc, alpha = .2)
#pl3d.plot_surface(xc,yc2,zc, alpha = .2)


#ax=pl.gca()
#ax.hold(True)
#print (yc)
pl.plot(xc,yc1,zc)
pl.plot(xc,yc2,zc)

#plotting grid 
pl.plot(xpoint,ypoint,'o')
pl.show()
```

```python
#these are the points the rays must go through all on xy plane
points = []
for i in range(0,len(xpoint)-1):
    points.append([xpoint[i],ypoint[i], 0])
#print (points)
points3 = np.array(points[1])
print ('These are the coordinates: ', points3)
```

```python
#making the rays from each point
d1 = np.array([.5,1,1]) #vector for incident ray
pl3d = pl.figure().gca(projection = '3d')
pl3d.set_xlim(left = -10, right = 10)
pl3d.set_ylim(bottom = -10, top = 10)
#pl3d.set_zlim(bottom=-.06, top = .06)
for i in range (0,len(points)):
    points2 = points[i]
    x = np.linspace(-3,3,10)
    y = ((x - points2[0])/(d1[0]))*d1[1] +points2[1]
    z = ((x - points2[0])/(d1[0]))*d1[2] + points2[2]
    
    pl3d.plot(x,y,z)
    ax=pl.gca()
    ax.hold(True)
pl.show()
    
```

```python
coeff1 = np.array([1,1,1,1]) #a,b,c,d for plane
#gradient of plane for reflecting
x,y,z = np.meshgrid(range(-10,10), range(-10,10), range(-1010))
U = coeff1[0]*x + coeff1[1]*y + coeff1[2]*z
dU = coeff1[:-1] #gradient (for plane)

#print (dU)
#plane
xp, yp = np.meshgrid(range(-10,10), range(-10,10))
zp = (-coeff1[0]*xp - coeff1[1]*yp + coeff1[3])/coeff1[2]

#normalize normal line
UNorm = np.sqrt((coeff1[0])**2 + (coeff1[1])**2 + (coeff1[2])**2)
N = ([u/UNorm for u in dU])


#plot of normal line?
origin = np.array([0,0,0]) #x_0,y_0,z_0 (for where to intersect)
t = (range(-10,10))
xL = origin[0] + t*coeff1[0]
yL = origin[1] + t*coeff1[1]
zL = origin[2] + t*coeff1[2]

pl3d = pl.figure().gca(projection = '3d')
pl3d.set_xlim(left = 10, right = -10)
pl3d.set_ylim(bottom = -10, top = 10)
pl3d.set_zlim(bottom = -10, top = 10)
pl3d.plot_surface(xp,yp,zp, alpha = .2)
pl3d.plot(xL,yL,zL, 'black')

for i in range (0,len(points)):
    points2 = points[i]
    #incident
    xi = np.linspace(-3,3,10)
    yi = ((xi - points2[0])/(d1[0]))*d1[1] +points2[1]
    zi = ((xi - points2[0])/(d1[0]))*d1[2] + points2[2]
    pl3d.plot(xi,yi,zi)
    
    #reflected
    N2 = [x*2 for x in N]
    d2 = d1 -N2

    xr = np.linspace(-3,3,10)
    yr = ((xr - points2[0])/(d2[0]))*d2[1] + points2[1]
    zr = ((xr - points2[0])/(d2[0]))*d2[2] + points2[2]
    pl3d.plot(xr,yr,zr)
    
    ax=pl.gca()
    ax.hold(True)

ax=pl.gca()
ax.hold(True)
pl.show()
```

```python
#doing the same reflection but with finite length and vector notation

coeff1 = np.array([1,1,1,1]) #a,b,c,d for plane
#gradient of plane for reflecting
x,y,z = np.meshgrid(range(-10,10), range(-10,10), range(-1010))
U = coeff1[0]*x + coeff1[1]*y + coeff1[2]*z
dU = coeff1[:-1] #gradient (for plane)

#print (dU)
#plane
xp, yp = np.meshgrid(range(-10,10), range(-10,10))
zp = (-coeff1[0]*xp - coeff1[1]*yp + coeff1[3])/coeff1[2]

#normalize normal line
UNorm = np.sqrt((coeff1[0])**2 + (coeff1[1])**2 + (coeff1[2])**2)
N = ([u/UNorm for u in dU])


#plot of normal line?
origin = np.array([0,0,0]) #x_0,y_0,z_0 (for where to intersect)
t = (range(-10,10))
xL = origin[0] + t*coeff1[0]
yL = origin[1] + t*coeff1[1]
zL = origin[2] + t*coeff1[2]

pl3d = pl.figure().gca(projection = '3d')
pl3d.set_xlim(left = 10, right = -10)
pl3d.set_ylim(bottom = -10, top = 10)
pl3d.set_zlim(bottom = -10, top = 10)
pl3d.plot_surface(xp,yp,zp, alpha = .2)
pl3d.plot(xL,yL,zL, 'black')

#plot of incident with finite length given point and vector
VectL = np.array([.5,1,1]) #direction
VNorm = np.sqrt(VectL[0]**2 + VectL[1]**2 + VectL[2]**2)
VectLNorm = ([u/VNorm for u in VectL])
VectLNorm = np.array(VectLNorm)

for i in range (0,len(points)):
    points2 = points[i]
    x = []
    y = []
    z = []
    for t in range (0,10):
        L = np.sqrt(((VectL[0]*t)**2 + (VectL[2]*t)**2 + (VectL[2]*t)**2))
        xL = points2[0] + t*VectLNorm[0]
        yL = points2[1] + t*VectLNorm[1]
        zL = points2[2] + t*VectLNorm[2]
        if L <= 20:
            x.append(xL)
            y.append(yL)
            z.append(zL)
    pl3d.plot(x,y,z)    



ax=pl.gca()
ax.hold(True)
pl.show()
```
