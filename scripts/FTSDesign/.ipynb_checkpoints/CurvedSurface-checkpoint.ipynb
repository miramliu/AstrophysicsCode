```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
```

elliptic paraboloid
$$\frac{x^2}{a^2}+\frac{y^2}{b^2} = \frac{z}{c}$$

```python
coeff3 = np.array([6,6,1]) #a,b,c sign of c determines if it is up or down!
xpe, ype = np.meshgrid(range(-10,10), range(-10,10))
#zpe = (((xpe**2)/(coeff3[0])**2) + ((ype**2)/(coeff3[1])**2))*coeff3[2]
zpe = ((xpe/coeff3[0])**2 + (ype/coeff3[1])**2)*coeff3[2]
```

```python
def zcoord(x,y):
    z = ((x/coeff3[0])**2 + (y/coeff3[1])**2)*coeff3[2]
    return z
```

```python
pl3d = pl.figure().gca(projection = '3d')
pl3d.plot_surface(xpe,ype,zpe)

ax=pl.gca()
ax.hold(True)
pl.show()
```

```python
#gradient of elliptic parabola
#First define U(x,y,z) = c
U = (xpe/coeff3[0])**2 + (ype/coeff3[1])**2 - (zpe/coeff3[2])
```

```python
#returns the gradient of U, which should be the unit vector of the norm at a certain point x,y,z (z is arbitrary?)
def dU(U,x,y):
    N = []
    dUx = 2*x*coeff3[2]/(coeff3[0]**2)
    dUy = 2*y*coeff3[2]/(coeff3[1]**2)
    dUz = -coeff3[2]
    N.append(dUx)
    N.append(dUy)
    N.append(dUz)
    UNorm = np.sqrt((N[0])**2 + (N[1])**2 + (N[2])**2)
    N = ([u/UNorm for u in N])
    return N
```

```python
#normalize a 3d array x
def normalized(x):
    xnorm = np.sqrt(x[0]**2 + x[1]**2 + x[2]**2)
    X = ([xx/xnorm for xx in x])
    return X
```

```python
#plot of normal line given point 
pointN = np.array([0,0,zcoord(0,0)]) #x_0,y_0,z_0
t = np.linspace(0,20,10)
N = (dU(U,pointN[0],pointN[1]))
N = np.array(N) #normalized unit vector

print (N)

xN = pointN[0] + t*N[0]
yN = pointN[1] + t*N[1]
zN = pointN[2] + t*N[2]
```

```python
#plot a line in 3d space? (incident ray)
# arbitrary vector direction
d1 = np.array([1,1,1])
d1 = normalized(d1)
print ('d1: ', d1)

#define it such that the point it passes through is pointN
t = np.linspace(0,10,20)
xLi = pointN[0] + t*(d1[0])
yLi = pointN[0] + t*(d1[1])
zLi = pointN[0] + t*(d1[2])

#determine the direction in 3d space of the reflected vector
c = np.dot(N,d1)
N2 = ([x*2*c for x in N])
d2 = [d1[0] + N2[0],d1[1] + N2[1],d1[2] + N2[2]]
    
#d2 = normalized(d2)
print('d2: ', d2)
print ('N2: ', N2)
```

```python
xLr = pointN[0] + t*(d2[0])
yLr = pointN[0] + t*(d2[1])
zLr = pointN[0] + t*(d2[2])

pl3d = pl.figure().gca(projection = '3d')

pl3d.plot(xN,yN,zN, 'white')
pl3d.plot(xLi,yLi,zLi, 'red')
pl3d.plot(xLr,yLr,zLr, 'yellow')
pl3d.plot_surface(xpe,ype,zpe)


ax=pl.gca()
ax.hold(True)
pl.show()
```
