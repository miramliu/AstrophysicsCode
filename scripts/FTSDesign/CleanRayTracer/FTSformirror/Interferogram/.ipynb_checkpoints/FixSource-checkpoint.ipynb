```python
def circle(c,r):
    x = np.linspace((c[0]-r),(c[0]+r),100)
    yp = []
    yn = []
    for i in range (100):
        y1 = np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        y2 = -np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        yp.append(y1)
        yn.append(y2)
    return x,yp,yn
x,y1,y2 = circle((160.375,-113),1.5875)
```
