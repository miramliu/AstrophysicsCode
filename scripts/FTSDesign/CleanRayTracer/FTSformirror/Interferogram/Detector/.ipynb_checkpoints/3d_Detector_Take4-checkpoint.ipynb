```python
def circlepoints(c,r):
    x = np.linspace((c[0]-r+.000000001),(c[0]+r-.000000001),10)
    points=[]
    for i in range (10):
        y1 = np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        y2 = -np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        point1 = [x[i],y1,c[2]]
        point2 = [x[i],y2,c[2]]
        points.append(point1)
        points.append(point2)
    return points
p = circlepoints([0,0,0],1)
for i in range(len(p)):
    pl.scatter(p[i][0], p[i][1])
pl.show()
```

```python
def ArrangedPoints(c,r):
    x = np.linspace((c[0]-r+.000000001),(c[0]+r-.000000001),10)
    points=[]
    for i in range (10):
        y1 = -np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        y2 = np.sqrt((r**2)-(x[i]-c[0])**2)+c[1]
        point1 = [x[i],y1,c[2]]
        if i <=5:
            y2 = c[1]+2*r
        if i >5:
            y2 = c[1]
        point2 = [x[i],y2,c[2]]
        points.append(point1)
        points.append(point2)
    return points
p = ArrangedPoints([0,0,0],1)
for i in range(len(p)):
    pl.scatter(p[i][0], p[i][1])
pl.show()
```

```python
def gaussian3d(x,y,sig,mu): #assuming is symmetric
    return (1/(np.sqrt(2*np.pi*(sig**2))*(sig**2)))*np.exp(-(((x-mu)**2)/(2*sig**2) + ((y-mu)**2)/(2*sig**2)))
#mu is the mean or expectation 
#sig is the standard deviation
```

```python
def fillcircle(r,center,num):
    ppr = int(num*.5) #number of sections needed per quarter circle#points (sections) per radius
    xstart = (center[0] - r)
    xend = (center[0] + r)
    xs = np.linspace(xstart,xend,ppr)
    ystart = (center[1]-r)
    yend = (center[1]+r)
    ys = np.linspace(ystart,yend,ppr)
    points = []
    for i in range(len(xs)):
        for j in range (len(ys)):
            if (xs[i]-center[0])**2 + (ys[j]-center[1])**2 <= r**2:
                p = [xs[i],ys[j]]
                points.append(p)
    return points
```
