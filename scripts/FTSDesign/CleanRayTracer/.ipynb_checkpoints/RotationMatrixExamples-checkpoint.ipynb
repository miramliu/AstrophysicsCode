```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
from RayTraceFunClean import *
from random import uniform
```

```python
Pi = [0,0,0]
Vi = [1,1,1]
Li = 10
x,y,z = ML(Pi,Vi,Li)
pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(x,y,z)
pl.show()
```

```python
#Rx (pos and neg)
Pi = [0,0,0]
Vi = [0,1,1]
Li = 10
x,y,z = ML(Pi,Vi,Li)

Vr = np.array(np.dot(Vi,Rx(np.pi/4)))
xr,yr,zr = ML(Pi, Vr[0],Li)

Vrn = np.array(np.dot(Vi,Rx(-np.pi/4)))
xrn,yrn,zrn = ML(Pi, Vrn[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(xr,yr,zr, color = 'seagreen') #positive
pl3d.plot(x,y,z, color = 'slateblue') #original
pl3d.plot(xrn,yrn,zrn, color = 'darkviolet') #negative

pl.show()
```

```python
#Rx (pos and neg) (SAME IF VECTOR IN OPPOSITE DIRECTION)
Pi = [0,0,0]
Vi = [0,-1,1]
Li = 10
x,y,z = ML(Pi,Vi,Li)

Vr = np.array(np.dot(Vi,Rx(np.pi/4)))
xr,yr,zr = ML(Pi, Vr[0],Li)

Vrn = np.array(np.dot(Vi,Rx(-np.pi/4)))
xrn,yrn,zrn = ML(Pi, Vrn[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(xr,yr,zr, color = 'seagreen') #positive
pl3d.plot(x,y,z, color = 'slateblue') #original
pl3d.plot(xrn,yrn,zrn, color = 'darkviolet') #negative

pl.show()
```

```python
#Ry (pos and neg)
Pi = [0,0,0]
Vi = [1,0,1]
Li = 10
x,y,z = ML(Pi,Vi,Li)

Vr = np.array(np.dot(Vi,Ry(np.pi/4)))
xr,yr,zr = ML(Pi, Vr[0],Li)

Vrn = np.array(np.dot(Vi,Ry(-np.pi/4)))
xrn,yrn,zrn = ML(Pi, Vrn[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(xr,yr,zr, color = 'seagreen')
pl3d.plot(x,y,z, color = 'slateblue')
pl3d.plot(xrn,yrn,zrn, color = 'darkviolet')

pl.show()
```

```python
#Rz (pos and neg)
Pi = [0,0,0]
Vi = [1,1,0]
Li = 10
x,y,z = ML(Pi,Vi,Li)

Vr = np.array(np.dot(Vi,Rz(np.pi/4)))
xr,yr,zr = ML(Pi, Vr[0],Li)

Vrn = np.array(np.dot(Vi,Rz(-np.pi/4)))
xrn,yrn,zrn = ML(Pi, Vrn[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(x,y,z, color = 'slateblue')
pl3d.plot(xr,yr,zr, color = 'seagreen')
pl3d.plot(xrn,yrn,zrn, color = 'darkviolet')

pl.show()
```

```python
#comparing Rxyz and individual rotation (SEEMS TO BE CORRECT)
Pi = [0,0,0]
Vi = [1,1,0]
Li = 10
x,y,z = ML(Pi,Vi,Li)
thet7 = [-.3104,0,-1.31067]
v3d = np.array(np.dot(Vi,Rxyz(thet7)))
x3d,y3d,z3d = ML(Pi, v3d[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(x,y,z, color = 'slateblue')
pl3d.plot(x3d,y3d,z3d, color = 'red')
pl.show()
```

```python
#Rz (pos and neg)
Pi = [0,0,0]
Vi = [-1,-1,0]
Li = 10
x,y,z = ML(Pi,Vi,Li)

Vr = np.array(np.dot(Vi,Rz(np.pi/4)))
xr,yr,zr = ML(Pi, Vr[0],Li)

Vrn = np.array(np.dot(Vi,Rz(-np.pi/4)))
xrn,yrn,zrn = ML(Pi, Vrn[0],Li)

Vrn2 = np.array(np.dot(Vrn[0],Rz(-np.pi/8)))
xrn2,yrn2,zrn2 = ML(Pi, Vrn2[0],Li)

pl3d = pl.figure().gca(projection = '3d')
pl3d.plot(x,y,z, color = 'slateblue')
pl3d.plot(xr,yr,zr, color = 'seagreen')
pl3d.plot(xrn,yrn,zrn, color = 'darkviolet')
pl3d.plot(xrn2,yrn2,zrn2, color = 'red')

pl.show()
```
