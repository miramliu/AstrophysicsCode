```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
from RayTraceFunCleaner import *
from random import uniform
import random
```

```python
def circularsource(r): #radius
    xpoint = []
    ypoint = []
    zpoint = []
    for x in range(-r,r):
        for y in range(-r,r):
            d = numpy.sqrt((x**2) + (y**2))
            if d <= r: 
                xpoint.append(x)
                ypoint.append(y)
                zpoint.append(int(0))
    return xpoint, ypoint, zpoint
```

```python
x,y,z=circularsource(8)
pl3d = pl.figure().gca(projection = '3d')
pl.plot(x,y,z, marker = 'o')
pl.show()
#circular source/generates points
```

```python
origin1 =  [-64.15,-128,0]
thet = [np.pi/2,0,.24553]
x1,y1,z1 = transform(x,y,z,origin1,thet)
pl3d = pl.figure().gca(projection = '3d')
pl.plot(x1,y1,z1, marker = 'o')
pl.show()
#supposedly the source for E1 
```

```python
#given sourcepoint = [x,y,z] where x etc is an array
def FS(specnum,sourcepoint,sourcethet,ellipseorigin):
    if type(sourcepoint[0]) is int or type(sourcepoint[0]) is float or type(sourcepoint[0]) is numpy.float64:
        v1 = spec(specnum)
        negorigin = negvect(ellipseorigin)
        v2 = transformV(v1,sourcethet)
        sp = transformP(sourcepoint,negorigin)
    else: 
        v1 = spec(specnum)
        negorigin = negvect(ellipseorigin)
        v2 = transformV(v1,sourcethet)
        sp = []
        for i in range (0,specnum):
            spx = random.choice(sourcepoint[0])
            spy = random.choice(sourcepoint[1])
            spz = random.choice(sourcepoint[2])
            sourcepoint2 = [spx,spy,spz]
            spT = transformP(sourcepoint2,negorigin)
            sp.append(spT)
    return sp,v2
```

```python
#attempting to use the new FS
sourcepoint1 = [x1,y1,z1] #from circular source and tranform
sourcethet = [np.pi/2,0,.24553]
sp1, v1 = FS(1000,sourcepoint1, sourcethet,origin1)
```

```python
pl3dsurf = pl.figure().gca(projection = '3d')
for i in range (0,len(v1)):
    x,y,z = ML(sp1[i], v1[i], 200)
    pl3dsurf.plot(x,y,z, color = 'orange')
pl.show()

#assuming this works
```
