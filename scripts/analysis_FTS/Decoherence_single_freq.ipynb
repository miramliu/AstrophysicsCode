```python
#functions for calculating weighted averages 

#this takes in a list of signal/weight tuples, where we can manually assign the weights
data = [(data1, 1), (data2, 1), (data3, 1), (data4, 1)]

weighted_avg = [0 for i in range(samples_requested)] # empty list that will hold the weighted average data

#calculate the weighted average of data_list, save the value in ret_list
def weighted_average(data_list, ret_list):
    sum_weights = 0
    for dp in data_list:
        data_n = dp[0]
        weight = dp[1]
        sum_weights += weight
        data_x_weight = data_n * weight
        ret_list += data_x_weight
    ret_list *= (1 / sum_weights)
    return ret_list

```
