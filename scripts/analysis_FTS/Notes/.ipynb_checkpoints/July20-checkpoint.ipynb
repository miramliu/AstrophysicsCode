```python
file=open( '../../../data/raw_data/20160707_1940_x_mirror_1in_aperture.pkl' , 'rb')
d=pickle.load(file)
file.close()

i = 8
Nsize = 2**i
dt=(1/(d['sample freq'])) #period
T1=dt*(Nsize) #full period
v=(d['speed'])
X = v*T1 #full distance
dx = dt*v #smallest amount of distance travelled
total_t = (d['scan time']) #how long it ran
    
total_s = (d['samples requested']) #number of samples 
startpt = ((total_s - Nsize)/2) #starting point


endpt = startpt + Nsize #ending point


df = 1/T1
f = df*np.arange(Nsize/2)+df/2.0
fFull = df*np.arange((Nsize/2) + 1)+df/2.0


y = (d['sig0F']) 
D = y[startpt:endpt]
#D = np.flipud(D)

a = d['delay0F']/v
t = a[startpt:endpt]

pl.plot(D)
```

It should be noted that this interferogram is extremely symmetric. Therefore Tau is going to be fairly small. Hopefully, soon there will be a run with much faster scan speeds, and therefore a more viable candidate for determining tau. 

```python
#model function convoluted with RC circuit and AC coupling function
def psi1(znot,q,p,T):
    Ns = 256
    z = t
    dz = z[2]-z[1]
    f = np.linspace(0,0.5/dz, np.floor(Ns/2))
    f_full = np.linspace(0,0.5/dz, np.floor(Ns/2) +1)
    #original function
    y0=-np.cos(2*np.pi*q*(z-znot))*(np.exp(-((z-znot)**2)/(2*p**2)))/(np.sqrt(2*np.pi)*p)
    yf0=np.fft.rfft(y0)
    K_det =(1/(1+(1j*2*np.pi*f_full*T))) #kernel from detector
    K_cc =(2*np.pi*f_full*.559)/(1 + (1j*2*np.pi*.559)) #kernel from AC coupling
    K_f = np.multiply(K_det,K_cc)
    y1=np.fft.irfft(yf0*K_f*2*np.pi)
    return t,y1
#znot = 0.012, q = 8.57718120805, p = .04
```

The plot below is the plot of the data and the model with values guessed by eye. Hopefully the numbers given by the following MCMC will be close to these. 

```python
pl.plot(t,D, color = 'black', label = 'data')
t,Model = psi1(-.015,9.4,0.035, 0.001)
pl.plot(t,Model*.0005, color = 'red', label = 'model')
pl.legend()
```

```python
#likelihood function
#why is it ln?
def lnlike(params, t, D):
    znot, q, p, T, a = params 
    t,Modelf = psi1(znot,q,p, T)
    Dev = -.5*(np.dot((D-Modelf*a),D-Modelf*a))
    Sig=0.0005
    return (Dev/Sig)-np.log(1/Sig) #sigma? if sigma is large it becomes shallower
```

```python
nll = lambda *args: -lnlike(*args) 
result = op.minimize(nll, [-.02, 8.57, .05, .02, .0002], args=(t,D))
znot1,q1,p1,T1,a1 = result["x"]
```

```python
#prior function
#again, why is it ln?
def lnprior(params):
    znot, q, p, T, a = params 
    if -.1 < znot < 0.1 and 1.0 < q < 15.0 and .01 < p < .1 and 0 < T < 0.1 and 0.0 < a < .0009:
        return 0.0
    return -np.inf 
    #set znot to some delay because then tau  must move to accomodate as opposed to znot doing so. 
    #center znot at zero. (-.00001<z<.000001)
    #
```

```python
#log probability function
def lnprob (params, t, D):
    lp = lnprior(params)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(params, t, D)
```

```python
ndim, nwalkers = 5,260
pos = [result["x"] + 10**(-4)*np.random.randn(ndim) for i in range(nwalkers)]
```

```python
sampler = emcee.EnsembleSampler(nwalkers,ndim,lnprob, args=(t,D))
```

```python
sampler.run_mcmc(pos, 4000); #number of steps
```

```python
walks=(sampler.chain)
print(np.shape(walks))

for nn in range(0,ndim):
    pl.plot(walks[0,:,nn])
    pl.plot(walks[1,:,nn])
    pl.plot(walks[2,:,nn])
    titles = ['znot','q','p','T','a']
    pl.title(titles[nn])
    pl.show() #sigma = .001
```

These are the paths of the 5000 steps of three of the walkers.

These walkers return these corner plots. Again tau is highly correlated with znot. To get the values and their uncertainties, we use histograms and cumulative plots. 

```python
#histogram of znot
pl.hist(samples[:,0], 100,)
pl.title('histogram of znot')
pl.axvline(x = -0.016625410219655516)
pl.axvline(x = -.017)
pl.axvline(x = -.0162)
pl.show()
print 'this is the mean: ', np.mean(samples[:,0])
```

Using the mean of the histogram, then using a cumulative plot which plots the probability that a value will be less than or equal to x, we can calculate uncertainty. 

The range is shown on the histogram above, and is determined by the plot below.

We chose 10%. Therefore 10% to the left and right of the mean on the cumulative plot determines the error range. 

This is shown below.

```python
#cumulative plot of znot
pl.hist(samples[:,0], 100, normed = 1,  cumulative = True)
pl.xlim(-.022, -.014)
pl.axvline(x = -0.016625410219655516)
pl.axhline(y = .465)
pl.axhline(y = .565)
pl.axhline(y = .365)

pl.axvline(x = -.017)
pl.axvline(x = -.0162)
pl.title('Cumulative of znot')
pl.show()

```

From this plot it can be deduced that the value for znot is -.0166, and the range is from -.017 to -.0162.
