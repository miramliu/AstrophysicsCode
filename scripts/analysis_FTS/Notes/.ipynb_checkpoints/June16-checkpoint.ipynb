```python
file=open( '../../../data/raw_data/20160428_1313_W1204_ref.pkl' , 'rb')
d=pickle.load(file)
file.close()

i = 8
Nsize = 2**i
dt=(1/(d['sample freq'])) #period
T1=dt*(Nsize) #full period
v=(d['speed'])
X = v*T1 #full distance
dx = dt*v #smallest amount of distance travelled
total_t = (d['scan time']) #how long it ran
    
total_s = (d['samples requested']) #number of samples 
startpt = ((total_s - Nsize)/2) #starting point


endpt = startpt + Nsize #ending point


df = 1/T1
f = df*np.arange(Nsize/2)+df/2.0
fFull = df*np.arange((Nsize/2) + 1)+df/2.0


y = (d['sig0F']) 
D = y[startpt:endpt]
D = np.flipud(D)

a = d['delay0F']/v
t = a[startpt:endpt]
```

```python
def psi1(z_not,q,p,T):
    Ns = 256
    z = t
    dz = z[2]-z[1]
    f = np.linspace(0,0.5/dz, np.floor(Ns/2))
    f_full = np.linspace(0,0.5/dz, np.floor(Ns/2) +1)
    y0=-np.cos(2*np.pi*q*(z-z_not))*(np.exp(-((z-z_not)**2)/(2*p**2)))/(np.sqrt(2*np.pi)*p)
    yf0=np.fft.rfft(y0)
    K_det =(1/(1+(1j*2*np.pi*f_full*T)))
    K_cc =(2*np.pi*f_full*30*10*(9))/(1 + (1j*2*np.pi*30*10**(9)))
    K_f = np.multiply(K_det,K_cc)
    y1=np.fft.irfft(yf0*K_f*2*np.pi)
    return z,y1
#z_not = 0.012, q = 8.57718120805, p = .04
```

```python
def findahat(P,T):
    x,Modelf = psi1(findznot(P,T),8.57718120805,P,T) #z_not, q, p, T
    Mult = np.dot(D,Modelf)
    Mult2 = np.dot(Modelf,Modelf)
    aHat=Mult/Mult2
    return aHat
```

```python
def findznot(P,T): #this is z_not written as a function of tau
    Znotvec = np.linspace(-.026,.026,150)
    MinzNot = np.zeros((len(Znotvec)))
    for kk in range (0,len(Znotvec)):
        z_not = Znotvec[kk]
        x,Modelf = psi1(z_not,8.57718120805,P, T)
        ModelMin = np.where(Modelf == Modelf.min())
        ModelMin = ModelMin[0]
        ModelMin = ModelMin[0]
        DataMin = np.where(D == D.min())
        DataMin = DataMin[0]
        DataMin = DataMin[0]
        MinzNot[kk] = np.abs(1-(float(ModelMin)/float(DataMin)))
    miin = np.where(MinzNot == MinzNot.min())
    miin = miin[0]
    miin = miin[0]
    BestzNot = Znotvec[miin]
    return BestzNot
```

```python
pvec = np.linspace(0.03,.08,40)
znotvec = np.linspace(-.05,.05,40)
Tvec = np.linspace(.01,.07,40)
q = 8.89
#the function findznot will be used as a constant when making corner plots
#p = .05, T = .02
```

We decided to try and see if the optimizations would work better with z_naught as a free parameter. This meant we had three free parameters, z_naught, p, and T. q we assumed was around 8.8. 
The function findznot is used as the prior for corner plot where it is used as a prior. T prior is .02, p prior is .05. 

We also attempted to incoporate the AC coupling function into our kernel, but are unsure of T2 at the moment. We assumed it was 30Ghz from mathematical calculation. 

The three plots below are the corner plots of chi-squared for the three free parameters. 

Mira

```python
#p vs z
ChiMat1= np.zeros((len(znotvec),len(pvec)))
for ii in range (0,len(znotvec)):
    for jj in range (0, len(pvec)):
        z = znotvec[ii]
        p = pvec[jj]
        T = .02
        x,Modelf = psi1(z,8.89,p, T)
        Dev = np.dot((D-Modelf*(findahat(p,T))),D-Modelf*(findahat(p,T)))
        ChiMat1[ii,jj] = Dev

levs=(np.linspace(0.001,.1,80))
pl.contourf(znotvec,pvec,(ChiMat1.T))
pl.xlabel('z')
pl.ylabel('p')
pl.title('p vs z')
#pl.xscale('log')
pl.colorbar()
pl.show()
```

```python
#T vs z
ChiMat2= np.zeros((len(znotvec),len(Tvec)))
for ii in range (0,len(znotvec)):
    for jj in range (0, len(Tvec)):
        z = znotvec[ii]
        p = .05
        T = Tvec[jj]
        x,Modelf = psi1(z,8.89,p, T)
        Dev = np.dot((D-Modelf*(findahat(p,T))),D-Modelf*(findahat(p,T)))
        ChiMat2[ii,jj] = Dev

levs=(np.linspace(0.001,.1,80))
pl.contourf(znotvec,Tvec,(ChiMat2.T))
pl.xlabel('z')
pl.ylabel('T')
pl.title('T vs z')
#pl.xscale('log')
pl.colorbar()
pl.show()
```

```python
#T vs p
ChiMat3= np.zeros((len(pvec),len(Tvec)))
for ii in range (0,len(pvec)):
    for jj in range (0, len(Tvec)):
        p = pvec[ii]
        T = Tvec[jj]
        x,Modelf = psi1(findznot(p,T),8.89,p, T)
        Dev = np.dot((D-Modelf*(findahat(p,T))),D-Modelf*(findahat(p,T)))
        ChiMat3[ii,jj] = Dev

levs=(np.linspace(0.001,.1,80))
pl.contourf(pvec,Tvec,(ChiMat3.T))
pl.xlabel('p')
pl.ylabel('T')
pl.title('T vs p')
#pl.xscale('log')
pl.colorbar()
pl.show()
```

From calculating the minimum of each of these plots, we get the optimal values for each plot. 

From p vs znot we get p = .0505, and znot = -.024, with T = .02

From T vs znot we get, with p = .05 and znot = -.021, T = .0146

From T vs p we get, p = .0517, with znot = -.026, T = .0223

```python
pl.plot(t,D, label = 'data', color = 'black')

x,Modelf1 = psi1(-.024,8.89,0.05,0.02) 
pl.plot(x,Modelf1*findahat(0.05,0.02), label = 'p=.05, T=.022, z=-.024', color='blue')

x,Modelf2 = psi1(-.021,8.89,0.05,0.014) #original q: 8.57718120805
pl.plot(x,Modelf2*findahat(0.05,0.014), label = 'p = .05, T =.014, z = -.021', color='green')

x,Modelf3 = psi1(-.026,8.89,0.051,0.0223) #original q: 8.57718120805
pl.plot(x,Modelf3*findahat(0.051,0.0223), label = 'p = .05, T =.014, z = -.021', color='red')


pl.xlim(-.2,.2)
pl.legend()
pl.show()
```

Clearly all three of the plots using the variables recieved from each corner plot are very close, so much so that they almost overlap.

This can be directly compared to the interferogram reached by using the function findznot, as opposed to treating it as a free parameter.
