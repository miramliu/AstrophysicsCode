This is a summary of the work done from July 16th to July 18th. It involved writing functions that find the intersection points of an ellipsoid and a line and then from that the tangent plane at that point. Given that, the correct reflecting vector can be found of an incident ray. (see pages 85-89)

An ellipse is made in the x-y plane and is then rotated around the x-axis. An example is below. The original ellipse (in blue) is rotated by $\frac{\pi}{2}$ to become the second ellipse (in green). 

Mira

```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
from raytracefunnotes import *
```

```python
#how this is rotated around the x axis by pi/2
pl3d = pl.figure().gca(projection = '3d')
x,y1,y2,z = createellipse([263.9,256,64.15])
pl3d.plot(x,y1,z)
x,y,z = rotatestrand(np.pi/2,[263.9,256,64.15] )
pl.plot(x,y,z ,'green')
pl.show()
```

The entire ellipsoid is made by doing this rotated $2\pi$ . A quarter of it is shown below. This is created by creating a series of points from rotating the ellipse small angles around the x axis and connecting them via triangulation to create a surface.

```python
pl3dsurf = pl.figure().gca(projection = '3d')
X,Y,Z = createz(50,np.pi/2, [263.9,256,64.15])
X.pop(0)
Y.pop(0)
Z.pop(0)
pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2, antialiased=True)
pl.title('1/4 Ellipsoid')
pl.show()
```

An example of these functions will be used with a unit sphere below. 

```python
#testing the ellipselineint function
#pli = [0,0,0] #(x0,y0,z0)
#v1 = [1,1,1] #a,b,c
#coeffellipse = [1,1,1]
xint1,yint1,zint1 = ellipselineint([0,0,0],[1,1,1],[1,1,1])
pl3dsurf = pl.figure().gca(projection = '3d')
X,Y,Z = createz(50,np.pi*1.99, [1,1,1])
X.pop(0)
Y.pop(0)
Z.pop(0)
pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2, antialiased=True)
pli = [[0,0,0]] #(x0,y0,z0)
vl = [1,1,1] #(a,b,c)
v11 = [-1,-1,-1]
Ll = int(3)
xline,yline,zline = make_line(pli[0],vl,Ll)
xline1,yline1,zline1 = make_line(pli[0],v11,Ll)
pl3dsurf.plot(xline,yline,zline, color = 'yellow')
pl3dsurf.plot(xline1,yline1,zline1, color = 'yellow')
xint,yint,zint = ellipselineint([0,0,0],[1,1,1],[1,1,1])
pl3dsurf.plot([xint[0]], [yint[0]], [zint[0]], marker = 'o', color = 'white')
pl3dsurf.plot([xint[1]], [yint[1]], [zint[1]], marker = 'o', color = 'white')
pl.title('sphere, ray, points of intersection')
pl.show()
```

Shown above is a sphere (blue) with a ray (yellow) going through it and the two points of intersection (white). From these two intersection points a tangent plane can be calculated. So, using one of the points and the equation of the ellipsoid, the tangent plane is created. This is shown below. 

```python
pl3dsurf = pl.figure().gca(projection = '3d')
xp1,yp1,zp1= tangplane([0,0,0],[1,1,1],[1,1,1], 2)
pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2, antialiased=True)
pl3dsurf.plot_surface(xp1,yp1,zp1, alpha = .2)
pl3dsurf.plot([xint[0]], [yint[0]], [zint[0]], marker = 'o', color = 'white')
pl.show()
```

This is the tangent plane (gray) at the intersection point (white). From this plane we can determine the reflecting angle and so find the ray.  

```python
#tangent plane
pli = [[0,0,0]] #(x0,y0,z0)
vl = [1,1,1] #(a,b,c)
Ll = int(300)
xline,yline,zline = make_line(pli[0],vl,Ll)
pl3dsurf = pl.figure().gca(projection = '3d')
X,Y,Z = createz(50,np.pi*1.99, [263.9,256,64.15])
X.pop(0)
Y.pop(0)
Z.pop(0)
pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2, antialiased=True)
pli = [0,0,0] #(x0,y0,z0)
v1 = [1,1,1] #a,b,c
coeffellipse = [263.9,256,64.15]
xint,yint,zint = ellipselineint(pli,v1,coeffellipse)
xp,yp,zp= tangplane(pli,v1,coeffellipse,300)
pl3dsurf.plot_surface(xp,yp,zp, alpha = .2)
pl.plot(xline,yline,zline, color = 'yellow')
pl3dsurf.plot([xint[0]],[yint[0]],[zint[0]], marker = 'o', color = 'white')
pl.show()
```

This is an example of the actual ellipsoid, a ray, its intersection point, and the tangent plane!
