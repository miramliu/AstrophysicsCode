```python
import numpy as np
import matplotlib.pyplot as pl
import itertools
from mpl_toolkits.mplot3d import Axes3D
from raytracefunnotes import *
```

```python
#two ellipses next to each other
coeffellipse = [263.9,256,64.15]
X,Y,Z = createzbounded(50,np.pi/4, coeffellipse, 64.15)
X.pop(0)
Y.pop(0)
Z.pop(0)

origin1 = [90,0]
X1,Y1,Z1 = createzboundshift(50,np.pi/4, coeffellipse, 64.15,origin1)
X1.pop(0)
Y1.pop(0)
Z1.pop(0)
```

```python
#now get a ray to reflect off of one of them, and a second ray to reflect off of the second one, 
r = int(100)
p = [0,200,-100] #(x0,y0,z0)
v = [0.01,1,.01] #(a,b,c)
L = int(200)
coeffellipse = [263.9,256,64.15]
xi,yi,zi,xr,yr,zr,xp,yp,zp  = reflectellipse(coeffellipse,r,v,p,L)
xint,yint,zint = ellipselineint(p,v,coeffellipse)
origin1 = [90,0]
xi1,yi1,zi1,xr1,yr1,zr1,xp1,yp1,zp1  = reflectellipseshift(coeffellipse,r,v,p,L, origin1)

pl3dsurf = pl.figure().gca(projection = '3d')
pl3dsurf.plot([xint[0]],[yint[0]],[zint[0]], marker = 'o', color = 'white')
#pl3dsurf.plot(xline,yline,zline, color = 'yellow')
pl3dsurf.plot(xi,yi,zi, color = 'yellow')
pl3dsurf.plot(xr,yr,zr, color = 'red')

xint1,yint1,zint1 = ellipselineintshift(p,v,coeffellipse, origin1)
pl3dsurf.plot([xint1[0]],[yint1[0]],[zint1[0]], marker = 'o', color = 'white')
pl3dsurf.plot(xi1,yi1,zi1, color = 'yellow')
pl3dsurf.plot(xr1,yr1,zr1, color = 'red')

pl3dsurf.plot_trisurf(X,Y,Z, linewidth = .2, antialiased=True)
pl3dsurf.plot_trisurf(X1,Y1,Z1, linewidth = .2, antialiased=True)
pl.show()

```
