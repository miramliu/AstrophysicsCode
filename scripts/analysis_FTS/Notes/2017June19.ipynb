```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl
%matplotlib inline
import pickle
import seaborn as sns
import sys
import numpy as np
import scipy.optimize as op
from scipy.optimize import curve_fit

sns.set_palette("PuBuGn_d",15)
sns.set_style("whitegrid")
#import emcee
```

```python
with open('../../../data/raw_data/20170615_1729_30mms_reference.pkl', 'rb') as f:
    d = pickle.load(f, encoding='latin1') 
```

```python
i = 9
Nsize = 2**i
dt=(1/(d['sample freq'])) #period
T1=dt*(Nsize) #full period
v=(d['speed'])
X = v*T1 #full distance
dx = dt*v #smallest amount of distance travelled
total_t = (d['scan time']) #how long it ran
    
total_s = (d['samples requested']) #number of samples 
startpt = ((total_s - Nsize)/2) #starting point


endpt = startpt + Nsize #ending point
startpt = int(startpt)
endpt = int(endpt)


df = 1/T1
f = df*np.arange(Nsize/2)+df/2.0
fFull = df*np.arange((Nsize/2) + 1)+df/2.0


y = (d['sig0F']) 
D = y[startpt:endpt]
#D = np.flipud(D)

a = d['delay0F']/v
t = a[startpt:endpt]

pl.plot(t,D)
pl.title('Interferogram at 30mm/s')
pl.show()
```

```python
#model function convoluted with RC circuit and AC coupling function
def psi1(znot,q,p,T):
    Ns = Nsize
    z = t
    dz = z[2]-z[1]
    f = np.linspace(0,0.5/dz, np.floor(Ns/2))
    f_full = np.linspace(0,0.5/dz, np.floor(Ns/2) +1)
    #original function
    y0=-np.cos(2*np.pi*q*(z-znot))*(np.exp(-((z-znot)**2)/(2*p**2)))/(np.sqrt(2*np.pi)*p)
    yf0=np.fft.rfft(y0)
    K_det =(1/(1+(1j*2*np.pi*f_full*T))) #kernel from detector
    K_cc =(2*np.pi*f_full*.559)/(1 + (1j*2*np.pi*.559)) #kernel from AC coupling
    K_f = np.multiply(K_det,K_cc)
    y1=np.fft.irfft(yf0*K_f*2*np.pi)
    return t,y1
#znot = 0.012, q = 8.57718120805, p = .04
```

This is then used in an MCMC to find the most appropriate constants in the hypothetical interferogram model. This interferogram model, or Model equation.

The Model equation we chose was 
$y_0=-cos{2\pi q(z-z_0)}e^-\frac{(z-z_0)^2}{2 p^2}\frac{1}{\sqrt{2\pi p}}$

This Model is clearly dependent on three variables; $z_0$, q, and p. $z_0$ shifts the Model on the x-axis. q determines the frequency of oscillation. p determines the width of the Model. 
However it is clear from the plot below (with $z_0$, q, and p chosen by eye) that there is also an important scale factor, which we called A. 

There is also a time constant involved due to convolution! This time decay contributes to the asymmetry of an interferogram.

Using an optimization function to recieve the closest values we get approximations of the four constants needed. The Model with these functions and the Data whose time constant we want is shown below. And MCMC is run to check these constants and ensure they are accurate.

```python
#VISUAL COMPARISON TO GUESS PARAMETERS BY EYE to put into optimals above
pl.plot(t,D+.001, label = 'Data', color = 'black')
x,Modelf = psi1(4.06765371e-03, 1.97765752e+01,   1.32557258e-02,   7.14676279e-05) #z,q,p,T

pl.plot(x,Modelf*1.32627361e-05, label = 'Model', color = 'red')
pl.xlim(-.2,.2)
pl.title('Data and MCMC Model')
pl.legend()
pl.savefig('WHEEEE')
pl.show()
```

Several of the corner plots were warped so I wrote another Model equation where $z_0$ is a constant. This gets rid of one free parameters hopefully leading to a closer approximation of the other parameters. This was also to account for the fact that $z_0$ and $\tau$ are functions of each other and so influence each other. From this we get the following function:


```python
#model function convoluted with RC circuit and AC coupling function
def psi2(q,p,T):
    Ns = Nsize
    z = t
    dz = z[2]-z[1]
    znot = 4.06765371e-03
    f = np.linspace(0,0.5/dz, np.floor(Ns/2))
    f_full = np.linspace(0,0.5/dz, np.floor(Ns/2) +1)
    #original function
    y0=-np.cos(2*np.pi*q*(z-znot))*(np.exp(-((z-znot)**2)/(2*p**2)))/(np.sqrt(2*np.pi)*p)
    yf0=np.fft.rfft(y0)
    K_det =(1/(1+(1j*2*np.pi*f_full*T))) #kernel from detector
    K_cc =(2*np.pi*f_full*.559)/(1 + (1j*2*np.pi*.559)) #kernel from AC coupling
    K_f = np.multiply(K_det,K_cc)
    y1=np.fft.irfft(yf0*K_f*2*np.pi)
    return t,y1
#znot = 0.012, q = 8.57718120805, p = .04
```
