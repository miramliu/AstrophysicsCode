```python
import numpy as np
from scipy.optimize import curve_fit
```

```python
Data = np.genfromtxt('../../data/raw_data/ChopperRun.txt')
```

```python
pl.plot(Data[:,0], Data[:,1])
```

```python
def func(x,a,b,c):
    return a * np.exp(-x*b)*(1-np.exp(-x*c))

Xdata = Data[:,0]
Ydata = Data[:,1]
```

```python
popt,pcov = curve_fit(func,Xdata,Ydata)
```

```python
print popt
#this means that the optimal tau is .049, when tau2 is .59
```

```python
print pcov
```

```python
perr = np.sqrt(np.diag(pcov))
print perr
print (perr/popt)*100
```

```python
pl.plot(Xdata,func(Xdata,popt[0],popt[1],popt[2]))
pl.plot(Xdata,Ydata, marker = 'o',ls='none',color='tomato')
```
