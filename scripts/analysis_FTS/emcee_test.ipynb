```python
import numpy as np
import matplotlib
import matplotlib.pyplot as pl
%matplotlib inline
import pickle
import seaborn as sns
import sys
import numpy as np
from scipy.optimize import curve_fit

sns.set_palette("coolwarm",15)
sns.set_style("whitegrid")

import emcee
```

```python
# Choose the "true" parameters.
m_true = -0.9594
b_true = 4.294
f_true = 0.534

# Generate some synthetic data from the model.
N = 50
x = np.sort(10*np.random.rand(N))
yerr = 0.1+0.5*np.random.rand(N)
y = m_true*x+b_true
y += np.abs(f_true*y) * np.random.randn(N)
y += yerr * np.random.randn(N)

yo = m_true*x+b_true
```

```python
pl.plot(x,y,'ro', color = 'gray')
pl.plot(x,yo)
```

A = np.vstack((np.ones_like(x), x)).T 
#np.vstack: stacks arrays in sequence vertically (by row)
#np.ones_like: returns an array of 1s, with the same shape and type of a given array
C = np.diag(yerr * yerr)
cov = np.linalg.inv(np.dot(A.T, np.linalg.solve(C, A)))
#np.linalg.inv calculates the multiplicative inverse of a matrix
#np.linalg.solve solves a system of linear equations
b_ls, m_ls = np.dot(cov, np.dot(A.T, np.linalg.solve(C, y)))

print 'b:', b_ls, 'm:', m_ls

y1 = m_ls*x+b_ls
pl.plot(x,yo, label = 'original', color = 'gray')
pl.plot(x,y1,label = 'model')
pl.legend()
pl.show()

```python
def lnlike(theta, x, y, yerr):
    m, b, lnf = theta
    model = m * x + b
    inv_sigma2 = 1.0/(yerr**2 + model**2*np.exp(2*lnf))
    return -0.5*(np.sum((y-model)**2*inv_sigma2 - np.log(inv_sigma2)))
```

```python
import scipy.optimize as op
nll = lambda *args: -lnlike(*args)
result = op.minimize(nll, [m_true, b_true, np.log(f_true)], args=(x, y, yerr))
m_ml, b_ml, lnf_ml = result["x"]
#print 'b: ', b_ml, 'm: ', m_ml
```

y2 = m_ml*x+b_ml

```python
def lnprior(theta):
    m, b, lnf = theta
    if -5.0 < m < 0.5 and 0.0 < b < 10.0 and -10.0 < lnf < 1.0:
        return 0.0
    return -np.inf
```

```python
def lnprob(theta, x, y, yerr):
    lp = lnprior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, x, y, yerr)
```

```python
ndim, nwalkers = 3, 100
pos = [result["x"] + 1e-4*np.random.randn(ndim) for i in range(nwalkers)]
sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob, args=(x, y, yerr))
sampler.run_mcmc(pos, 500);
```

```python
samples = sampler.chain[:, 50:, :].reshape((-1, ndim))
```
