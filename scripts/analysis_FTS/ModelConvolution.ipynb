```python
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
Ns = 256
z = np.linspace(-.4,.4,Ns)
dz = z[2]-z[1]
f = np.linspace(0,0.5/dz, np.floor(Ns/2))
f_full = np.linspace(0,0.5/dz, np.floor(Ns/2) +1)
q=8.57718120805
p=0.04
T = .03
z_not = 0.012

y0=-np.cos(2*np.pi*q*(z-z_not))*(np.exp(-((z-z_not)**2)/(2*p**2)))/(np.sqrt(2*np.pi)*p)
pl.plot(z,y0)
pl.xlim(-.4,.4)
```

```python

K = T*np.exp(-(z-z_not)/T)
pl.plot(K)
```

```python
yf0=np.fft.rfft(y0)
K_f = T/(1+(1j*2*np.pi*f_full*T))
y1=np.fft.irfft(yf0*K_f*2*np.pi)
pl.plot(z,y1, '--r', label = 'convoluted output')
pl.plot(z,y0, label = 'original input')
#pl.xlim(-.4,.4)
pl.axhline(.1)
```
