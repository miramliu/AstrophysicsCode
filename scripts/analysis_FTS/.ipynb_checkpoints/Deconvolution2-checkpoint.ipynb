```python
#spectrum
i = 13
Nsize = 2**i
dt=(1/(d['sample freq'])) #period
t=dt*np.arange(Nsize) #full period
v=(d['speed'])
x = v*t #full distance
dx = dt*v #smallest amount of distance travelled
total_t = (d['scan time']) #how long it ran
    
total_s = (d['samples requested']) #number of samples 
startpt = ((total_s - Nsize)/2) #starting point
endpt = startpt + Nsize #ending point
    
F = (d['sig0F']) 
A = np.hanning(Nsize)*F[startpt:endpt] #certain section of signal 
S = np.fft.rfft(A) 

a = d['delay0F']
b = a[startpt:endpt]
    
s = S[:-1]
u = np.abs(S)
dNu = 1/(Nsize*dx) #1/ number of points times smallest distance travelled (length of x axis)
Nu = dNu*np.arange(Nsize/2) # .5number of points times length of x-axis
NuFull = dNu*np.arange((Nsize/2) + 1)

tau = .05 #ms
k_v = 1/((1/tau)+1j*2*np.pi*300*NuFull) #kernel *fourier transform of e^-t/Tau

n = 3 #greater number filters out more high frequencies
sig = n/(tau*(2*np.pi)) 
            
f = (1/(sig*np.sqrt(2*np.pi)))*np.exp(-(NuFull**2)/(2*sig**2)) #gaussian
B = f*np.fft.rfft(A)/k_v*tau #fourier transform of interferogram times gaussian divided by kernel
c = np.fft.irfft(B) #inverse fourier transform of deconvolved spectrum

pl.plot (300*Nu,np.abs(s), label = 'convoluted spectrum') 
pl.plot(300*NuFull, np.abs(np.fft.rfft(c))/k_v*tau, label = 'deconvoluted spectrum')

#L=np.sum(np.abs(np.fft.rfft(c))/k_v*tau)
#pl.plot(300*NuFull, (1/L)*np.abs(np.fft.rfft(c))/k_v*tau, label = 'normalized spectrum')


axes = pl.gca()
axes.set_xlim([0,600])
axes.set_ylim([0,12])
pl.legend()
pl.show()
```

```python
g = (np.abs(np.fft.rfft(c))/k_v*tau)[:2**7] #REASONABLE FREQUENCY RANGE, ONLY 0 - 600
f = 300*NuFull[:2**7]
L=np.sum(g) #area under spectrum
#pl.plot(300*NuFull, (1/L)*np.abs(np.fft.rfft(c))/k_v*tau, label = 'normalized spectrum')

pl.plot(300*NuFull[:2**7],(1/L)*(np.abs(np.fft.rfft(c))/k_v*tau)[:2**7] )

#axes = pl.gca()
axes.set_xlim([0,600])
axes.set_ylim([0,1])
pl.legend()
pl.show()
```
